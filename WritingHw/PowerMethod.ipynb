```python
import numpy as np
```

```python
def PowerMethod(R, X, times):
    '''
    :param R: Matrix to be solved.
    :param X: Vector at the beginning.
    :param times: Times to be executed.
    :return: None
    '''
    for i in range(times):
        print(f"Times: {i}, X_{i}: {X} \n")
        X = R @ X        
```

```python
def StdPowerMethod(R, X, times):
    '''
    standardized power method
    :param R: Matrix to be solved.
    :param X: Vector at the beginning.
    :param times: Times to be executed.
    :return: None
    '''
    Y = X / np.absolute(X).max()
    for i in range(times+1):
        print(f"Times: {i}, X_{i}: {X} \n \t\t\tY_{i}: {Y}")       
        X = R @ Y
        Y = X / np.absolute(X).max()    
```

```python
R = np.array([[-4.,4.,1.], [0.,1.,0.], [0.,0.,1.]])
X = np.array([1.,2.,3.])
PowerMethod(R,X,15)
```

```python
StdPowerMethod(R,X,15)
```

```python
A = np.array([[0, -0.5], [1./3, 5./6]])
X1 = np.array([1., 1.])
StdPowerMethod(A, X1, 30)
```

```python
PowerMethod(A,X1,30)
```
